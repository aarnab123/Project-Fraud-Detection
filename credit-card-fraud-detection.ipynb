```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['NormalizedAmount'] = scaler.fit_transform(df["Amount"].values.reshape(-1,1))
df.drop(columns=['Amount','Time'],inplace=True)
X=df.drop(columns=['Class'],axis=1)
Y=df['Class']
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,Ytrain,Ytest=train_test_split(X,Y,test_size=0.3,random_state=45)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
DT=DecisionTreeClassifier()
RC=RandomForestClassifier(n_estimators=150)
```

```python
DT.fit(Xtrain,Ytrain)
Y_DT_pr=DT.predict(Xtest)
score_dt=DT.score(Xtest,Ytest)
RC.fit(Xtrain,Ytrain)
Y_RC_pr=RC.predict(Xtest)
score_rc=RC.score(Xtest,Ytest)
print('Decission Tree Score is: ',score_dt*100)
print('Random forest Score is: ',score_rc*100)
```

```python
Dt_table=pd.DataFrame(np.c_[Ytest,Y_DT_pr],columns=['Original','Predicted'])
Rf_table=pd.DataFrame(np.c_[Ytest,Y_RC_pr],columns=['Original','Predicted'])
Dt_table.head()
```

```python
Rf_table.head()
```

```python
from sklearn.metrics import accuracy_score, precision_score, confusion_matrix, recall_score,r2_score,classification_report
```

```python
Ac_DT=accuracy_score(Ytest,Y_DT_pr)
print('Accuracy Score of Decission Tree: ',Ac_DT)
print('R2 Score of Decission Tree :',r2_score(Ytest,Y_DT_pr))
print('Precision Score of Decision Tree: ',precision_score(Ytest,Y_DT_pr))
print(confusion_matrix(Ytest,Y_DT_pr))
Ac_RFC=accuracy_score(Ytest,Y_RC_pr)
print('Accuracy Score of Random Forest : ',Ac_RFC)
print('R2 Score of Random Forest :',r2_score(Ytest,Y_RC_pr))
print('Precision Score of Random Forest: ',precision_score(Ytest,Y_RC_pr))
print(confusion_matrix(Ytest,Y_RC_pr))
```

**Extra Tree Classifier**

```python
from sklearn.ensemble import ExtraTreesClassifier
ET=ExtraTreesClassifier(n_estimators=500,random_state=15)
ET.fit(Xtrain,Ytrain)
Y_ET_pr=ET.predict(Xtest)
print(classification_report(Ytest,Y_ET_pr))
```

```python
Ac_ETC=accuracy_score(Ytest,Y_ET_pr)
print('Accuracy Score from Extra-Tree-Classifier is:',Ac_ETC)
print(confusion_matrix(Ytest,Y_ET_pr))
```

**Bagging Clasifier**

```python
from sklearn.ensemble import BaggingClassifier
Bac=BaggingClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=150)
Bac.fit(Xtrain,Ytrain)
Y_Bac_pr=Bac.predict(Xtest)
print(classification_report(Ytest,Y_Bac_pr))
```

```python
Ac_Bac=accuracy_score(Ytest,Y_Bac_pr)
print('Accuracy Score is: ',Ac_Bac)
print(confusion_matrix(Ytest,Y_Bac_pr))
```

**K Nearest neighbours**

```python
from sklearn.neighbors import KNeighborsClassifier
KNN=KNeighborsClassifier(n_neighbors=7)
KNN.fit(Xtrain,Ytrain)
Y_KNN_pr=KNN.predict(Xtest)
print(classification_report(Ytest,Y_KNN_pr))
```

```python
Ac_KNN=accuracy_score(Ytest,Y_KNN_pr)
print('Accuracy Score is: ',Ac_KNN)
print(confusion_matrix(Ytest,Y_KNN_pr))
```

**SVM**

```python
from sklearn.svm import SVC
SVM=SVC()
SVM.fit(Xtrain,Ytrain)
Y_SVM_pr=SVM.predict(Xtest)
print(classification_report(Ytest,Y_SVM_pr))
```

```python
Ac_SVM=accuracy_score(Ytest,Y_SVM_pr)
print('Accuracy Score is: ',Ac_SVM)
print(confusion_matrix(Ytest,Y_SVM_pr))
```

**Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
LR=LogisticRegression()
LR.fit(Xtrain,Ytrain)
Y_Lr_pr=LR.predict(Xtest)
print(classification_report(Ytest,Y_Lr_pr))
```

```python
Ac_lr=accuracy_score(Ytest,Y_Lr_pr)
print('Accuracy Score is: ',Ac_lr)
print(confusion_matrix(Ytest,Y_Lr_pr))
```

```python
Accuracy_table=pd.DataFrame({'Classifiers':['DecisionTree','RandomForest',
                                            'ExtraBaggingClassifier','BaggingClassifier','KNN',
                                            'SVM','LogisticRegression'],
                             'Accuracy':[Ac_DT,Ac_RFC,Ac_ETC,Ac_Bac,Ac_KNN,
                                         Ac_SVM,Ac_lr]})
Accuracy_table.sort_values(ascending=False,by='Accuracy')
```
